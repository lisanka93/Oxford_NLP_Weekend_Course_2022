## What is ML?
Machine learning is a subfield of artificial intelligence, which is broadly defined as the capability of a machine to imitate intelligent human behavior. Artificial intelligence systems are used to perform complex tasks in a way that is similar to how humans solve problems.

Machine learning is one way to use AI. It was defined in the 1950s by AI pioneer Arthur Samuel as “the field of study that gives computers the ability to learn without explicitly being programmed.”

Michael Shulman, a lecturer at MIT Sloan and head of machine learning at Kensho, compared the traditional way of programming computers, or “software 1.0,” to baking, where a recipe calls for precise amounts of ingredients and tells the baker to mix for an exact amount of time. Traditional programming similarly requires creating detailed instructions for the computer to follow.

But in some cases, writing a program for the machine to follow is time-consuming or impossible, such as training a computer to recognize pictures of different people. While humans can do this task easily, it’s difficult to tell a computer how to do it. Machine learning takes the approach of letting computers learn to program themselves through experience. 

Machine learning starts with data — numbers, photos, or text, like bank transactions, pictures of people or even bakery items, repair records, time series data from sensors, or sales reports. The data is gathered and prepared to be used as training data, or the information the machine learning model will be trained on. The more data, the better the program.

From there, programmers choose a machine learning model to use, supply the data, and let the computer model train itself to find patterns or make predictions. Over time the human programmer can also tweak the model, including changing its parameters, to help push it toward more accurate results.

## Bias and Variance

When developing a machine learning model two common problems may occur - **underfitting** and **overfitting**.
Underfitting means that the model isn't able to sufficiently identify the relation between input and output values. The common reasons for this are the model simplicity and too small a training set. On the other hand, overfitting is a situation when a model is too strongly related to a training set resulting in bad performance against unseen data. The perfect model will be ideally between these two situations, being able to generalize to unseen data well.

## Explanatory Data Analysis (EDA)

### Missing Data
There's no imputing in NLP. Imputing can make sense in some cases with a numerical value (even then it should be used cautiously), but in general text is too diverse (unstructured data) for the concept of "most frequent text" to make any sense. 

When dealing with numerical values, there are several ways to deal with missing values:
- use the mean
- use the median
- use ML to predict them
- drop them (rows if only a few entries missing and columns if many entries are missing)

### Outliers
Outliers are also a phenomenon in numerical values, but should be mentioned given it is an important topic in ML. Outliers can be detected by, for example, visualising the data or by calculating the standard deviation and inspecting the items that fall outside.

In the boxplot below I have plotted the age of athletes from a dataset on olympic medals. We can clearly see an outliers, which was a typo or formatting issue (290 instead of 29).




### Unbalanced Data
Imbalanced classification is defined by a dataset with a skewed class distribution. Highly unbalanced data is a problem because your model might fail to generalise on the minority class and struggle distinguishing it from the majority class. The model might simply always predict the majority class.



#### Dealing with unbalanced data

There are several ways to deal with unbalanced data which include:
- **Undersampling**: 
Undersampling can be defined as removing some observations of the majority class. This is done until the majority and minority class is balanced out. Undersampling can be a good choice when you have a ton of data -think millions of rows. But a drawback to undersampling is that we are removing information that may be valuable.
- **Oversampling**: Oversampling can be defined as adding more copies to the minority class. Oversampling can be a good choice when you don’t have a ton of data to work with. A con to consider when undersampling is that it can cause overfitting and poor generalization to your test set
- **Artificial Data**
     - For numerical values: Synthetic Minority Oversampling Technique (SMOTE)
     - For textual data: one idea would be to translate the texts of the minority class into a different language and then translate it back into the original language



Note that for numerical data there is usually much more EDA and certain preprocessing you can (have to) do (e.g. checking for correlated features, feature scaling) but this is out of the scope of this course, given we are dealing with textural data.

## Validation Techniques - how do I know that my model performs well?

In order to know whether our model performs well we need to test it on unseen data (data which was not seen during training the model). Hence, we need to split our data into multiple parts.

Data splitting is the process of splitting data into 3 sets:

    - Data which we use to design our models (Training set)
    - Data which we use to refine our models (Validation set)
    - Data which we use to test our models (Testing set)

If we do not split our data, we might test our model with the same data that we use to train our model.

For example, if the model is a trading strategy specifically designed for Apple stock in 2008, and we test its effectiveness on Apple stock in 2008, of course it is going to do well. We need to test it on 2009’s data. Thus, 2008 is our training set and 2009 is our testing set.

Ok, so what is a validation set? The validation set is a set of data that we did not use when training our model that we use to assess how well these rules perform on new data. It can then be used to tune hyperparameters (parameters that are not learned by the model but are set by us) and input features for our model so that it gives us what we think is the best performance possible for new data.

We will talk more about hyperparameters later, when we look at different ML models, but some "hyperparameters" we have already seen in the previous notebook - like n-gram range (strictly speaking these are features but still). Depending on the n-gram range the model will perform better or worse on the validation set and will allow us to judge what the ideal n-gram range would be to maximise model performance.



But what is our validation test is not very representative of the whole data? There is an infinite number of possible combinations of train-test sets, however, we only experimented with one. How do we know that the very splitting we obtained is the most representative one? Maybe, if we had a different composition of train and test set, we would have a very different result.

A better way to validate our model is to use **cross-validation**. The idea of cross-validation arises because of the caveat explained above. It basically wants to guarantee that the score of our model does not depend on the way we picked the train and test set. It works as follows. It splits our dataset into K-folds, then the model is trained on K-1 folds and tested on the remaining one, for K iterations. So each time, because of the K rotations of the test set, the model is trained and tested on a new composition of data.



## Evaluation Metrics
- Accuracy
- Precision & Recall
- F1 Score

After training a model, we need a metric to understand how well (or not) it is performing on unseen data. In the case of supervised learning where we use labeled data, we can compare the model outcome with the human-produced label, also called **gold label**. Labels predicted by the model we will call **system labels**. 

Okay, let's say we create a spam detector that given a text, classifies it as spam or ham. To measure the performance of such a model, we could create a simple metric that tells us the percentage of the correct predictions (where a system label is identical to to a gold label). This metric is called **accuracy** and is expressed as a ratio of the number of correctly predicted system labels to the total number of guesses.

 There are 4 possible situations:
- The message was indeed spam and the system classified it as spam (**true positive**)
- The message was indeed spam and the system classified it as not spam (**false negative**)
- The message was not spam and the system classified it as not spam (**true negative**)
- The message was not spam and the system classified it as spam (**false positive**)

#### Accuracy
Using the terms above, **accuracy** is defined as the **percentage of correct predictions** for the test data. It can be calculated easily by dividing the number of correct predictions by the number of total predictions (the sum of true positives and true negatives divided by all samples). If we had a test set of 100 samples and the classifier got 80 of them correct, then the accuracy is 80%.

However, remember we talked about unbalanced datasets? Accuracy is not a meaningful metric for unbalanced datasets. Imagine a dataset with 2 classes, containing 100,000 tweets, from which only a small sample of tweets will are in the minority class (1% or so). Now, what will be the accuracy of this model if it **doesn't work** and always says that a sample belongs to the majority class? Let's calculate it. We have 100,000 tweets in the test set:
- 1,000 tweets are in the minority class
- 99,000 tweets are in the majority class

Since the model always says that a tweet is in the majority class, we have 1,000 false negatives and 99,000 true negatives. The accuracy is 99,000 / 100,000, which is 99%! 


#### Precision & Recall
To detect situations like this there are two other metrics, which are used together to create a useful set of evaluation metrics. 
The first one is called **precision** - it expresses the ratio of true positives to everything system labeled as positive (true positives and false positives).
The other one is called **recall** - it expresses the ratio of true positives to all samples which were gold labeled as positive (true positives and false negatives).

The metrics are calculated for each class, so for the minority class in the example above, precision and recall would both be 0.

Let's go through another example to demonstrate precision and recall: Let's assume we created a search engine and upon a user query, the serach engine returns 30 pages, only 20 of which are relevant, while failing to return 40 additional relevant pages. It's precision is 20/30 (20 out of all predictions were precise), while recall is 20/60 (only 20 out of all possible pages were returned).

#### F1-Score

It can get tedious to always carry 2 metrics around, so often, when reading research papers, you will come accross a metric called F1-Score. The F-score is a way of combining the precision and recall of the model, and it is defined as the harmonic mean of the model’s precision and recall.

### Transforming categorical labels into numerical values

```python
# some models in sklearn needs lables in numerical values, so let's use LabelEncoder
le = LabelEncoder()
bbc_data["Label"] = le.fit_transform(bbc_data["Category"])
```

### Setting a baseline for further experiments
A baseline is a method that uses heuristics, simple summary statistics, randomness, or machine learning to create predictions for a dataset. You can use these predictions to measure the baseline's performance (e.g., accuracy)-- this metric will then become what you compare any other machine learning algorithm against.

### Naive Bayes (NB) 
Naive-Bayes is a supervised learning probabilistic classifier. One of the most common applications of NB was spam detection. It is based on applying Bayes' probability theorem - the occurrence of an event impacts the probability of another event. The priors for each class and likelihoods for each word (more on that tomorrow) of the training data are calculated and then applied to new, unseen data The word "naive" is used to indicate the naive assumption of the conditional independence between samples' features (basically, it assumes there is no relation between word occurrences). Naive Bayes is a very fast classifier compared to Logistic Regression since it doesn't need to tune weights and parameters (just calculate the priors and likelihoods). Generally, because of the naive assumption, it is not used in serious applications, but sometimes may work quite well - as seen below.

```python
#creating our train test split - will ignore hold-out set for now since it is just for demo purposes
X_train, X_test, y_train, y_test = train_test_split(
    bbc_data['Text'],
    bbc_data['Label'],
    test_size=0.2,
    stratify = bbc_data['Label'],
    random_state=50,
)
```

```python
#instantiate countvectorizer 
count_vector = CountVectorizer(binary=True) 

#fit training data
training_data = count_vector.fit_transform(X_train)

#transform test data
testing_data = count_vector.transform(X_test)

naive_bayes = MultinomialNB()
naive_bayes.fit(training_data,y_train)

y_pred = naive_bayes.predict(testing_data)
```

```python
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average='weighted')
```

```python
conf_mat = confusion_matrix(y_test, y_pred)
conf_mat
```

```python
fig = plot_confusion_matrix(naive_bayes, testing_data, y_test, display_labels=naive_bayes.classes_)
fig.figure_.suptitle("Confusion Matrix for BBC Dataset")
plt.show()
```

The beautify of scikit-learn is that we can plug and play with different classifiers. So let's try a few more.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn import tree
```

### KNN Classifier
K-Nearest Neighbors (KNN) is one of the simplest algorithms used in Machine Learning for regression and classification problem. KNN algorithms use data and classify new data points based on similarity measures (e.g. distance function). Classification is done by a majority vote to its neighbors.



```python
knn = KNeighborsClassifier(n_neighbors=5)  #default neighbours is 5 - this is a hyperparameter which needs to be tuned. n_neighbors = 
knn.fit(training_data,y_train)

y_pred = knn.predict(testing_data)
accuracy_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average='weighted')
```

### Logistic Regression
Logistic regression is one of the most common classifiers (along with the NB). Logistic regression classifier develops a clever function that finds the dependency between input and output values. It uses a set of developed features, which are related with weights telling which feature is important for which class. Adjusting weights is done in a series of training and testing when the algorithm tries to find the best possible weights. Unlike the NB, this classifier requires a training stage, however, predicts generally better on more complex datasets. We will see how it works in more detail tomorrow.

```python
lr = LogisticRegression()#max_iter=1000)
lr.fit(training_data,y_train)

y_pred = lr.predict(testing_data)
accuracy_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average='weighted')
```

### Decision Tree
A Decision Tree is an algorithm, where the data is continuously split according to a certain parameter(s) - it is a simple decision-making model. Although Decision Trees are comfortable to inspect (since it is possible to simply plot the tree and see how it works), they are sensitive to variance and can be easily overfitted.

```python
tclf = tree.DecisionTreeClassifier()#max_depth = 10)  #to avoid DT overfitting, better to reduce max_depth
tclf.fit(training_data,y_train)
tree.plot_tree(tclf)
y_pred = tclf.predict(testing_data)

accuracy_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average='weighted')
```

### Multilayer Perceptron (simple neural network)
A multilayer perceptron (MLP) is a class of feedforward artificial neural network (ANN). An MLP consists of at least three layers of nodes: an input layer, a hidden layer and an output layer. Except for the input nodes, each node is a neuron that uses a nonlinear activation function. MLP utilizes a supervised learning technique called backpropagation for training.



```python
mlp = MLPClassifier(hidden_layer_sizes=(10,), activation='logistic', alpha=1e-4,
                    solver='sgd', tol=1e-4, random_state=1,
                    learning_rate_init=.01, verbose=True)

mlp.fit(training_data,y_train)

y_pred = mlp.predict(testing_data)
accuracy_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average='weighted')
```
